```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("compressed_data.csv")
```

```python
df.head()
```

```python
df.columns
```

# checking missing values 

```python
print(df.isnull().sum())
```

```python
df.info()
```

# handling missing values 

```python
df['last review'] = pd.to_datetime(df['last review'], errors= 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()}, inplace= True)
```

```python
df.dropna(subset = ['NAME', 'host name'] ,inplace = True)
```

```python
print (df.isnull().sum())
```

```python
df= df.drop(columns=['license','house_rules'], errors='ignore')
```

```python
df.head()
```

```python
# remove dollar sign and convert to float 

df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

# remove duplicates 

```python
df.drop_duplicates(inplace = True) 
```

```python
df.info()
```

```python
# descriptive statistics 
df.describe()
```

# what is the distribution of listing prices ? 

```python
plt.figure(figsize =(10,6 ))
sns.histplot(df['price'], bins=50 ,kde = True,color = 'blue')
plt.title('Distribution of listing price ')
plt.xlabel('price$')
plt.ylabel('frequency')
plt.show()
```

```python
summary :
The histogram shows a fairly even distribution of listing prices accross different 
price range accross different price ranges ,indicating no particular concentration
of listing in any specific price range. 
The KDE lines helps visulaize this even spread more clearly,confirming that dataset 
contains listings with a wide variety of prices .
```

# how are different room types  distributed ?

```python
df['room type']
```

```python
plt.figure(figsize= (8,5))
sns.countplot(x= 'room type',data = df,color = 'hotpink')
plt.title('room type distribution')
plt.xlabel('room type')
plt.ylabel('count')
plt.show ()
```

# how are listing distribution accross different neighbourhood ?

```python
plt.figure(figsize = (12,8))
sns.countplot(y= 'neighbourhood group',data= df,color = 'lightgreen', order= df['neighbourhood group'].value_counts().index)
plt.title('Number of Listing By Neighbouhood Group ')
plt.xlabel('count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

# what is the relation between price and room type ? 

```python
 plt.figure(figsize=(10,6))
sns.boxplot(x='room type', y= 'price',hue = 'room type', data = df, palette='Set1')
plt.title('price vs. Room Type ')
plt.xlabel('Room Type')
plt.ylabel('price($)')
plt.legend(title='Room Type')
plt.show()
```

# how has the number of reviews change over with time ?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of reviews over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show ()
```
